```python
import matplotlib.pyplot as plt
from matplotlib.axes import Axes
from matplotlib.patches import Patch
import seaborn as sns
import shap
from lime.lime_tabular import LimeTabularExplainer
from sklearn.calibration import calibration_curve
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, roc_auc_score, average_precision_score, roc_curve, confusion_matrix, precision_recall_curve
from sklearn.utils import shuffle
from sklearn.base import clone
from sklearn.neural_network import MLPClassifier
from catboost import CatBoostClassifier
import pandas as pd
import numpy as np
from scipy import stats
import os
from tqdm import tqdm, trange
import joblib


# Set Seaborn style
sns.set_style("darkgrid")

def save_plot_as_image(file_path, formats=['jpg', 'svg', 'png', 'pdf'], dpi=300):
	for fmt in formats:
		plt.savefig(f"{file_path}.{fmt}", format=fmt, dpi=dpi, bbox_inches='tight')
	plt.show()
	plt.close()

def find_closest_sample_to_mean(a: np.ndarray):
	mean_a = np.mean(a, axis=0).reshape(1,-1)
	rep_mean_a = np.repeat(mean_a, a.shape[0], axis=0)
	dists = np.linalg.norm(a - rep_mean_a, axis=1)
	min_dist_idx = np.argmin(dists)
	return a[min_dist_idx]

def evaluate_robustness(X, y, classifiers):
	metrics = ['accuracy_score', 'precision_score', 'recall_score', 'f1_score']

	if not os.path.exists('reports2/robustness_results.csv'):
		results = []

		for seed in trange(100, desc="Evaluating robustness"):
			np.random.seed(seed)
			X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)

			X_train_dict = {nl: X_train + np.random.normal(0, nl, X_train.shape) for nl in noise_levels}
			X_train_dict[0.0] = X_train

			for name in classifiers:
				for noise_level in noise_levels:
					X_noisy = X_train_dict[noise_level]
					X_noisy, y_noisy = shuffle(X_noisy, y_train, random_state=seed)

					ckpt_path = f'checkpoints/{name}_{int(noise_level * 100)}% Noise_{seed}.joblib'
					if os.path.exists(ckpt_path):
						classifier = joblib.load(ckpt_path)
					else:
						clf = {
							'Bagging': BaggingClassifier(random_state=seed),
							'MLP Classifier': MLPClassifier(max_iter=1000, random_state=seed),
							'CatBoost': CatBoostClassifier(verbose=0, random_state=seed),
						}[name]
						classifier = clf.fit(X_noisy, y_noisy)
						joblib.dump(classifier, ckpt_path)

					for metric in metrics:
						results += [{
							'Classifier': name,
							'Noise Level': noise_level,
							'Seed': seed,
							'Metric': metric,
							'Value': eval(f'{metric}(y_val, classifier.predict(X_val))'),
						}]

		results = pd.DataFrame(results)
		results.to_csv('reports2/robustness_results.csv', index=False)
	else:
		results = pd.read_csv('reports2/robustness_results.csv')

	for name in classifiers:
		fig, axes = plt.subplots(2, 4, figsize=(12, 6), sharey=True, height_ratios=[1, 0.25])
		for metric, ax in zip(metrics, axes[0, :]):
			sns.boxplot(data=results[(results['Classifier'] == name) & (results['Metric'] == metric)], 
			   			x='Noise Level', 
						y='Value', 
						gap=0.5,
						ax=ax,
						color='skyblue')
			ax.set_xticklabels([f'{int(nl * 100)}%' for nl in noise_levels])
			ax.set_title(metric.replace('_', ' ').capitalize())
		
		for metric, ax in zip(metrics, axes[1, :]):
			filtered = results[(results['Classifier'] == name) & (results['Metric'] == metric)]
			pivot_table = filtered.pivot_table(index='Noise Level', columns='Seed', values='Value').T\
				.describe().round(2)\
				.rename(columns={col: f'{int(col * 100)}%' for col in noise_levels})\
				.drop(index=['count', 'min', 'max'])\
				.rename(lambda s: s.capitalize(), axis=0)

			rcolors = plt.cm.BuPu(np.full(len(pivot_table.index), 0.1))
			ccolors = plt.cm.BuPu(np.full(len(pivot_table.columns), 0.1))

			table = ax.table(pivot_table, 
							 loc='center', 
							 cellLoc='center', 
							 rowLoc='center', 
							 colWidths=[0.2]*len(pivot_table.columns),
							 rowColours=rcolors,
							 colColours=ccolors,
							 edges='horizontal',
						)
			table.auto_set_font_size(False)
			for (i, j), cell in table.get_celld().items():
				if i == 0 or j == -1:
					cell.set_text_props(weight='bold', fontsize=8)
				else:
					cell.set_text_props(weight='normal', fontsize=7.5)
			ax.axis('off')
		fig.suptitle('Robustness Analysis')
		fig.tight_layout()
		save_plot_as_image(f'reports2/{name}_robustness_analysis')

def plot_roc_curve(classifier_name, X, y):
	results = []

	for seed in trange(100, desc="Evaluating ROC curve"):
		_, X_val, _, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)

		for noise_level in noise_levels:
			ckpt_path = f'checkpoints/{classifier_name}_{int(noise_level * 100)}% Noise_{seed}.joblib'
			classifier = joblib.load(ckpt_path)
			
			y_pred_proba = classifier.predict_proba(X_val)[:, 1]
			fpr, tpr, _ = roc_curve(y_val, y_pred_proba)

			fpr_grid = np.linspace(0, 1, 100)
			tpr_interp = np.interp(fpr_grid, fpr, tpr)

			for f, t in zip(fpr_grid, tpr_interp):
				results += [{
					'Noise Level': noise_level,
					'Seed': seed,
					'FPR': f,
					'TPR': t,
					'AUC': roc_auc_score(y_val, y_pred_proba),
				}]

	results = pd.DataFrame(results)

	plt.figure(figsize=(8,5))
	for noise_level, group in results.groupby("Noise Level"):
		fpr_grouped = group.groupby('FPR')
		mean_tpr = fpr_grouped['TPR'].mean()
		std_tpr = fpr_grouped['TPR'].std()
		mean_auc = group['AUC'].mean()
		std_auc = group['AUC'].std()
		
		fpr_values = mean_tpr.index

		plt.plot(fpr_values, mean_tpr, label=f'{noise_level*100:.0f}% Noise (AUC$ = {mean_auc:.2f} \pm  {std_auc:.2f}$)')
		plt.fill_between(fpr_values,
							mean_tpr - std_tpr,
							mean_tpr + std_tpr,
							alpha=.2)

	plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Chance')
	plt.xlabel('False Positive Rate')
	plt.ylabel('True Positive Rate')
	plt.title(f'ROC Curve')
	plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
	plt.grid(True, linestyle='--', alpha=0.7)
	plt.tight_layout()
	save_plot_as_image(f'reports2/{classifier_name}_roc_curve')

def plot_comparison_roc_curves(noise_level, X, y):
	results = []
	
	plt.figure(figsize=(8,5))
	for classifier_name in ['Bagging', 'MLP Classifier', 'CatBoost']:
		for seed in trange(100, desc="Evaluating ROC curve"):
			_, X_val, _, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)

			ckpt_path = f'checkpoints/{classifier_name}_{int(noise_level * 100)}% Noise_{seed}.joblib'
			classifier = joblib.load(ckpt_path)
			
			y_pred_proba = classifier.predict_proba(X_val)[:, 1]
			fpr, tpr, _ = roc_curve(y_val, y_pred_proba)

			fpr_grid = np.linspace(0, 1, 100)
			tpr_interp = np.interp(fpr_grid, fpr, tpr)

			auc = roc_auc_score(y_val, y_pred_proba)
			for f, t in zip(fpr_grid, tpr_interp):
				results += [{
					'Classifier': classifier_name,
					'Seed': seed,
					'FPR': f,
					'TPR': t,
					'AUC': auc,
				}]
	results = pd.DataFrame(results)
	for classifier_name, group in results.groupby("Classifier"):
		fpr_grouped = group.groupby('FPR')
		mean_tpr = fpr_grouped['TPR'].mean()
		std_tpr = fpr_grouped['TPR'].std()
		mean_auc = group['AUC'].mean()
		std_auc = group['AUC'].std()
		
		fpr_values = mean_tpr.index

		plt.plot(fpr_values, mean_tpr, label=f'{classifier_name} (AUC$ = {mean_auc:.2f} \pm  {std_auc:.2f}$)')
		plt.fill_between(fpr_values,
							mean_tpr - std_tpr,
							mean_tpr + std_tpr,
							alpha=.2)
	plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Chance')
	plt.xlabel('False Positive Rate')
	plt.ylabel('True Positive Rate')
	plt.title(f'{noise_level * 100:.0f}% Noise ROC Curve Comparison')
	plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
	plt.grid(True, linestyle='--', alpha=0.7)
	plt.tight_layout()
	save_plot_as_image(f'reports2/comparison_roc_curve')

def plot_pr_curve(classifier_name, X, y):
	results = []

	for seed in trange(100, desc="Evaluating PR Curve"):
		_, X_val, _, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)

		for noise_level in noise_levels:
			ckpt_path = f'checkpoints/{classifier_name}_{int(noise_level * 100)}% Noise_{seed}.joblib'
			classifier = joblib.load(ckpt_path)
			
			y_pred_proba = classifier.predict_proba(X_val)[:, 1]
			precision, recall, _ = precision_recall_curve(y_val, y_pred_proba)

			recall_grid = np.linspace(0, 1, 100)
			precision_interp = np.interp(recall_grid, recall[::-1], precision[::-1])

			for r, p in zip(recall_grid, precision_interp):
				results += [{
					'Noise Level': noise_level,
					'Seed': seed,
					'Precision': p,
					'Recall': r,
					'AUC': average_precision_score(y_val, y_pred_proba),
				}]

	results = pd.DataFrame(results)

	plt.figure(figsize=(8,5))
	for noise_level, group in results.groupby("Noise Level"):
		recall_grouped = group.groupby('Recall')
		mean_precision = recall_grouped['Precision'].mean()
		std_precision = recall_grouped['Precision'].std()
		mean_auc = group['AUC'].mean()
		std_auc = group['AUC'].std()
		
		recall_values = mean_precision.index

		plt.plot(recall_values, mean_precision, label=f'{noise_level*100:.0f}% Noise (AUC$ = {mean_auc:.2f} \pm  {std_auc:.2f}$)')
		plt.fill_between(recall_values,
							mean_precision - std_precision,
							mean_precision + std_precision,
							alpha=.2)

	plt.plot([0, 1], [1, 0], 'k--', lw=2, label='Random Chance')
	plt.xlabel('Recall')
	plt.ylabel('Precision')
	plt.title(f'PR Curve')
	plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
	plt.grid(True, linestyle='--', alpha=0.7)
	plt.tight_layout()
	save_plot_as_image(f'reports2/{classifier_name}_precision_recall_curve')

def plot_comparison_pr_curves(noise_level, X, y):
	results = []

	plt.figure(figsize=(8,5))
	for classifier_name in ['Bagging', 'MLP Classifier', 'CatBoost']:
		for seed in trange(100, desc="Evaluating PR Curve"):
			_, X_val, _, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)

			ckpt_path = f'checkpoints/{classifier_name}_{int(noise_level * 100)}% Noise_{seed}.joblib'
			classifier = joblib.load(ckpt_path)
			
			y_pred_proba = classifier.predict_proba(X_val)[:, 1]
			precision, recall, _ = precision_recall_curve(y_val, y_pred_proba)

			recall_grid = np.linspace(0, 1, 100)
			precision_interp = np.interp(recall_grid, recall[::-1], precision[::-1])

			auc = average_precision_score(y_val, y_pred_proba)
			for r, p in zip(recall_grid, precision_interp):
				results += [{
					'Classifier': classifier_name,
					'Seed': seed,
					'Precision': p,
					'Recall': r,
					'AUC': auc,
				}]

	results = pd.DataFrame(results)

	for classifier_name, group in results.groupby("Classifier"):
		recall_grouped = group.groupby('Recall')
		mean_precision = recall_grouped['Precision'].mean()
		std_precision = recall_grouped['Precision'].std()
		mean_auc = group['AUC'].mean()
		std_auc = group['AUC'].std()
		
		recall_values = mean_precision.index

		plt.plot(recall_values, mean_precision, label=f'{classifier_name} (AUC$ = {mean_auc:.2f} \pm  {std_auc:.2f}$)')
		plt.fill_between(recall_values,
							mean_precision - std_precision,
							mean_precision + std_precision,
							alpha=.2)

	plt.plot([0, 1], [1, 0], 'k--', lw=2, label='Random Chance')
	plt.xlabel('Recall')
	plt.ylabel('Precision')
	plt.title(f'{noise_level * 100:.0f}% Noise PR Curve Comparison')
	plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
	plt.grid(True, linestyle='--', alpha=0.7)
	plt.tight_layout()
	save_plot_as_image(f'reports2/comparison_precision_recall_curve')

def plot_confusion_matrices(classifier_name, X, y):
	fig, axes = plt.subplots(1, len(noise_levels), figsize=(15, 4))

	for ax, noise_level in zip(axes, noise_levels):
		cm = []
		for seed in trange(100, desc="Evaluating Confusion Matrix"):
			_, X_val, _, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)
			ckpt_path = f'checkpoints/{classifier_name}_{int(noise_level * 100)}% Noise_{seed}.joblib'
			classifier = joblib.load(ckpt_path)
			y_pred = classifier.predict(X_val)		
			cm += [confusion_matrix(y_val, y_pred)]

		cm = np.sum(cm, axis=0)

		sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, ax=ax)
		ax.set_title(f'{int(noise_level * 100)}% Noise')
		ax.set_xlabel('Predicted Label')
		ax.set_ylabel('True Label')
	
	fig.suptitle(f'Confusion Matrix ({cm.sum()} Samples)')
	fig.tight_layout()
	save_plot_as_image(f'reports2/{classifier_name}_confusion_matrix')

def get_best_seed(classifier_name, X, y):
	best_seed = None
	best_mean_accuracy = float('-inf')
	best_std_accuracy = float('inf')

	for seed in trange(100, desc="Selecting best model"):
		np.random.seed(seed)
		X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)

		X_train_dict = {nl: X_train + np.random.normal(0, nl, X_train.shape) for nl in noise_levels}
		X_train_dict[0.0] = X_train

		accuracies = [
			accuracy_score(
				y_val, 
				joblib.load(f'checkpoints/{classifier_name}_{int(noise_level * 100)}% Noise_{seed}.joblib')
					.predict(X_val)
			) 
			for noise_level in noise_levels
		]

		mean_accuracy = np.mean(accuracies)
		std_accuracy = np.std(accuracies)
		if mean_accuracy > best_mean_accuracy or (mean_accuracy == best_mean_accuracy and std_accuracy < best_std_accuracy):
			best_mean_accuracy = mean_accuracy
			best_std_accuracy = std_accuracy
			best_seed = seed
	
	with open(f'reports2/{classifier_name}_best_seed.txt', 'w+') as f:
		f.write(f'Best seed for {classifier_name}: {best_seed} (accuracy = {best_mean_accuracy:.2f} +/- {best_std_accuracy:.2f})')
	return best_seed

def plot_shap_summary(classifier_name: str, seed: int, df: pd.DataFrame, X: np.ndarray, y: np.ndarray):
	X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)

	wavelengths = np.array(list(map(float, df.drop('Label', axis=1).columns.tolist())))

	wavelength_ranges = np.linspace(wavelengths.min(), wavelengths.max(), 6)
	wavelength_bins = np.digitize(wavelengths, bins=wavelength_ranges)

	Xtrain_summary = shap.kmeans(X_train, 5)
	min_value, max_value = float('inf'), -float('inf')
	fig, axes = plt.subplots(1, 5, sharey=True, figsize=(25, 4))
	for noise_level, ax in zip(noise_levels, axes):
		ckpt_path = f'checkpoints/{classifier_name}_{int(noise_level * 100)}% Noise_{seed}.joblib'
		classifier = joblib.load(ckpt_path)
		explainer_shap = shap.KernelExplainer(classifier.predict_proba, Xtrain_summary)
		
		shap_values = explainer_shap.shap_values(X_val)[:, :, 1]

		mean_shap_values_per_column = np.abs(shap_values).mean(axis=0)

		shap_values_binned = {'names': [], 'values': []}
		for i in range(len(wavelength_ranges)):
			try:
				shap_values_binned['names'] += [f'[{wavelength_ranges[i]:.1f}, {wavelength_ranges[i+1]:.1f})']
				shap_values_binned['values'] += [mean_shap_values_per_column[wavelength_bins == (i+1)].sum()]
			except IndexError:
				shap_values_binned['names'][-1] = shap_values_binned['names'][-1][:-1] + ']'
				shap_values_binned['values'][-1] += mean_shap_values_per_column[wavelength_bins == (i+2)].sum()

		shap_values_binned['values'] = np.array(shap_values_binned['values'])
		
		max_value = max(max_value, shap_values_binned['values'].max())
		min_value = min(min_value, shap_values_binned['values'].min())

		bars = ax.barh(shap_values_binned['names'][::-1], list(shap_values_binned['values'])[::-1])
		if noise_level == 0:
			ax.set_yticklabels(shap_values_binned['names'][::-1], fontsize=8)
		for bar in bars:
			x, y, w, h = bar.get_x(), bar.get_y(), bar.get_width(), bar.get_height()
			ax.text(x=w, y=y + h / 2, s=f'  {w:.2f}', ha='left', va='center')

		ax.set_title(f'{int(noise_level * 100)}% Noise')
		ax.set_xlabel('')

	min_value, max_value = min_value - .1, max_value + .1
	for ax in axes:
		ax.set_xlim((min_value, max_value))
		ax.set_xticklabels(ax.get_xticklabels(), fontsize=8)
	
	fig.suptitle(f'SHAP Summary')
	fig.align_xlabels()
	fig.tight_layout()
	save_plot_as_image(f'reports2/{classifier_name}_shap_summary')

def plot_lime_explanation(classifier_name: str, seed: int, df: pd.DataFrame, X: np.ndarray, y: np.ndarray):
	X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=seed)

	explainer = LimeTabularExplainer(
		X_train,
		feature_names=df.drop('Label', axis=1).columns,
		class_names=['Class 0', 'Class 1'],
		discretize_continuous=True,
		random_state=42,
	)

	X_val_closest_sample_to_mean = find_closest_sample_to_mean(X_val)
	max_value = -float('inf')
	fig, axes = plt.subplots(5, 3, figsize=(8, 10), width_ratios=[1, 4, 2])
	for i, noise_level in enumerate(noise_levels):
		ckpt_path = f'checkpoints/{classifier_name}_{int(noise_level * 100)}% Noise_{seed}.joblib'
		classifier = joblib.load(ckpt_path)

		exp = explainer.explain_instance(
			X_val_closest_sample_to_mean,
			classifier.predict_proba,
			num_features=5,
			labels=[0, 1]
		)

		features = exp.as_list(label=1)
		features.sort(key=lambda x: x[1])
		
		feature_names = [f[0] for f in features]
		values = [f[1] for f in features]
		max_value = max(max_value, np.abs(values).max())

		class0_color, class1_color = 'deepskyblue', 'orangered'
		colors = [class0_color if v < 0 else class1_color for v in values]

		ax1: Axes = axes[i, 0]
		probs = exp.predict_proba
		ax1.bar_label(ax1.bar([''], probs[0], label='Class 0', color=class0_color, width=.1), label_type='center', fmt='%.2f', fontsize=7, weight='semibold')
		ax1.bar_label(ax1.bar([''], probs[1], label='Class 1', color=class1_color, bottom=probs[0], width=.1), label_type='center', fmt='%.2f', fontsize=7, weight='semibold')
		if i == 0: ax1.set_title(f'Prediction\nProbabilities', pad=12, fontdict={'fontsize': 10, 'weight': 'bold'})
		ax1.set_ylabel(f'{int(noise_level * 100)}% Noise', labelpad=12, fontdict={'fontsize': 10, 'weight': 'bold'})
		ax1.set_xticks([])
		ax1.set_yticks([])
		ax1.set_facecolor('white')
		
		ax2: Axes = axes[i, 1]
		ax2.bar_label(ax2.barh(feature_names, values, color=colors), label_type='center', fmt='%.3f', fontsize=7, weight='semibold')
		ax2.vlines([0], -1, len(values), colors='black', linestyles='-', lw=.5)
		ax2.text(0, len(values)-.25, '   Class 1', ha='left', va='bottom', fontsize=10, color=class1_color)
		ax2.text(0, len(values)-.25, 'Class 0   ', ha='right', va='bottom', fontsize=10, color=class0_color)
		ax2.grid(True)
		if i == 0: ax2.set_title(f'LIME Explanations', pad=12, fontdict={'fontsize': 10, 'weight': 'bold'})

		ax3: Axes = axes[i, 2]
		table_data = pd.DataFrame([list(X_val_closest_sample_to_mean)], columns=df.drop('Label', axis=1).columns)
		table_data = table_data[[f.split(' ')[2] for f in feature_names]].T
		table_data.rename({0: 'Value'}, axis='columns', inplace=True)
		table_data.insert(0, 'Feature', table_data.index)
		table_data['Rank'] = np.arange(0, len(table_data))
		table_data.set_index('Rank', inplace=True)
		table_data.sort_index(ascending=False, inplace=True)
		table_data['Value'] = table_data['Value'].apply(lambda x: f'{x:.2f}')
		
		ax3.table(cellText=table_data.values, 
					colLabels=table_data.columns, 
					cellColours=[[colors[::-1][i] for c in r] for i, r in enumerate(table_data.values.tolist())],
					cellLoc='center', 
					loc='center')
		if i == 0: ax3.set_title(f'Top-5 Feature Values', pad=12, fontdict={'fontsize': 10, 'weight': 'bold'})
		ax3.set_xticks([])
		ax3.set_yticks([])
		ax3.set_facecolor('white')

	max_value *= 1.1
	for ax in axes[:, 1]:
		ax.set_aspect('auto')
		ax.set_xlim(-max_value, max_value)
		ax.set_ylim(-1, 5)
		ax.set_xticks(np.concatenate((-np.arange(0, max_value, .02)[::-1], np.arange(0, max_value, .02))))
		ax.set_xticklabels(ax.get_xticklabels(), fontsize=7)
		ax.set_yticklabels(ax.get_yticklabels(), fontsize=6)
			
	fig.tight_layout()
	save_plot_as_image(f'reports2/{classifier_name}_lime_explanation')

noise_levels = [0.0, 0.05, 0.1, 0.15, 0.2]

os.makedirs('reports2', exist_ok=True)
os.makedirs('checkpoints', exist_ok=True)

df = pd.read_excel('data2/preprocessed_combined_917.xlsx')
X = df.drop('Label', axis=1).values
y = df['Label'].values


classifiers = ['Bagging', 'MLP Classifier', 'CatBoost']

evaluate_robustness(X, y, classifiers)
plot_comparison_pr_curves(0.1, X, y)
plot_comparison_roc_curves(0.1, X, y)

for name in classifiers:
	print(f'Classifier: {name}')
	plot_roc_curve(name, X, y)
	plot_pr_curve(name, X, y)
	plot_confusion_matrices(name, X, y)
	best_seed = get_best_seed(name, X, y)
	plot_shap_summary(name, best_seed, df, X, y)
	plot_lime_explanation(name, best_seed, df, X, y)
```
